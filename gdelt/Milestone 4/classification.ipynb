```python
test_article_list = pd.DataFrame.from_dict(test_article)

# convert label to a numerical variable
test_article_list['label_num'] = test_article_list.category.map({'murder':0, 'theft':1, 'corruption':2, 'terrorist':3,  'violence':4})
```

```python
test_article_list
```

```python
X = test_article_list.summary
y = test_article_list.label_num

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
vect = CountVectorizer()

vect.fit(X_train)
X_train_dtm = vect.transform(X_train)
X_test_dtm = vect.transform(X_test)
```

## Naive Bayes

```python
nb = MultinomialNB()
```

```python
%time nb.fit(X_train_dtm, y_train)
```

```python
# make class predictions for X_test_dtm
y_pred_class = nb.predict(X_test_dtm)
```

```python
metrics.accuracy_score(y_test, y_pred_class)
```

```python
metrics.confusion_matrix(y_test, y_pred_class)
```
