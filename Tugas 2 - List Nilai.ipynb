
## LATIHAN 2 - Tugas List Nilai

Diberikan sekumpulan Nilai Random antara 0 sampai dengan 100 sebanyak 250 nilai. Buatlah program untuk melakukan hal berikut:

Tampilkan Angka Ganjil dari ke 250 nilai tersebut
Tampilkan berapa banyak yang mendapatkan nilai lebih dari 70
Tampilkan standar deviasi dari sekumpulan nilai tersebut
Tampilkan nilai maksimum dan minimum
Tampilkan modus dari sekumpulan nilai tersebut

```python
from statistics import *
import random
import math

# GENERATE RANDOM
nilaiTugas = []
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))

print(nilaiTugas)

nilaiGanjil = []
nilai70 = []
maksimum = 0
minimum = 100
modus = 0
frekuensiModus = 0
total = 0
selisihKuadrat = 0
rata2 = 0

for n in nilaiTugas:
    # Check bilangan Ganjil
    if(n%2 == 1):
        nilaiGanjil.append(n)
        
    # Check bilangan lebih dari 70
    if(n > 70):
        nilai70.append(n)
        
    # Check bilangan maksimum
    if(n > maksimum):
        maksimum = n
    
    # Check bilangan minimum
    if(n < minimum):
        minimum = n
    
    # Menghitung total
    total += n

#  Menghitung rata-rata
rata2 = total/len(nilaiTugas)

# Menghitung jumlah selisih nilai tugas dengan rata-rata
for n in nilaiTugas:
    # Mengjumlahkan selisih kuadrat
    # Menggunakan pow karena masalah konversi nilai
    selisih = n - rata2
    selisihKuadrat = math.pow(selisih,2)

# Menghitung Standar Deviasi
v = selisihKuadrat/len(nilaiTugas)
standarDeviasi = math.sqrt(v)

# Menghitung Modus
# Sorting
i = 0
while i < len(nilaiTugas):    
    j = i+1
    while j < len(nilaiTugas):
        if nilaiTugas[i] > nilaiTugas[j]:
            temp = nilaiTugas[i]
            nilaiTugas[i] = nilaiTugas[j]
            nilaiTugas[j] = temp
        j += 1
    i += 1

#  Mencari Modus
angkaSama = False
tempModus = 0
tempFrekuensiModus = 0

i = 0
while i < len(nilaiTugas): 
    if not angkaSama:
        tempModus = nilaiTugas[i]
        tempFrekuensiModus = 1
        angkaSama = True
        i += 1
    else:
        if tempModus == nilaiTugas[i]:
            tempFrekuensiModus += 1
            if tempFrekuensiModus > frekuensiModus:
                frekuensiModus = tempFrekuensiModus
                modus = tempModus
            i += 1
        else:
            angkaSama = False    

print("\n# Nilai Ganjil", nilaiGanjil)
print("\n# Nilai lebih dari 70", nilai70)
print("\n# Nilai Standar Deviasi", standarDeviasi)
print("\n# Nilai Maksimum", maksimum)
print("\n# Nilai Minimum", minimum)
print("\n# Nilai Modus", modus, "sejumlah", frekuensiModus,"kali muncul")
```
