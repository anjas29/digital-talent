## Tugas Analisa Data

### Data: Penduduk miskin yang menderita penyakit kronis
Dataset ini berisi data mengenai jumlah individu dengan status kesejahteraan 30% terendah yang menderita penyakit kronis per Provinsi

Penjelasan mengenai Variabel pada Dataset ini:

kode_provinsi: Kode Provinsi
nama_provinsi: Nama Provinsi
kelompok_usia: Kelompok Usia (0 - < 15 tahun, 15 - < 45 tahun, 45 - 60 tahun)
jenis_kelamin: Jenis Kelamin
jumlah_individu: Jumlah Individu yang Menderita Penyakit Kronis pada Kelompok Usia dan Jenis Kelamin tersebut

Sumber: https://data.go.id/dataset/penduduk-miskin-yang-menderita-penyakit-kronis

```python
import pandas as pd

data_penyakit_kronis = pd.read_csv('data_penyakit_kronis.csv', delimiter = ',', encoding = "ISO-8859-1")
data_penyakit_kronis.head()
```

## 2. Statistik Data

```python
# Mean
data_mean = data_penyakit_kronis['jumlah_individu'].mean()
print('Mean :', data_mean)

# Median
data_median = data_penyakit_kronis['jumlah_individu'].median()
print('Median :', data_median)

# Modus
data_modus = data_penyakit_kronis['jumlah_individu'].mode()
print('Modus :', data_modus[0])

# deskripsi data
data_penyakit_kronis['jumlah_individu'].describe()
```

## 3. Visualisasi
Visualisasi menggunakan Diagram Batang dengan mengelompokkan data berdasarkan kelompok usia dan jenis kelamin.

```python
data_penyakit_kronis.groupby(["kelompok_usia","jenis_kelamin"]).sum()['jumlah_individu']
```

```python
ax = data_penyakit_kronis.groupby(["kelompok_usia", "jenis_kelamin"]).sum()['jumlah_individu']
ax = ax.plot(kind='bar', title ="Jumlah Penderita Penyakit Kronis", figsize=(15,15), legend=True, fontsize=12)
ax.set_xlabel("Kategori", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```
